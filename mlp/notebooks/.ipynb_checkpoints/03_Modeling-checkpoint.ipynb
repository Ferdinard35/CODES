
## Healthcare Fraud Detection


---
## 1. Setup & Load Clean Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    ConfusionMatrixDisplay, RocCurveDisplay, classification_report
)

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 120

DATA_DIR  = r'C:\Users\Ferdinard\Desktop\CODES\mlp\data'
PLOTS_DIR = r'C:\Users\Ferdinard\Desktop\CODES\mlp\plots'
os.makedirs(PLOTS_DIR, exist_ok=True)

# Load clean datasets saved on Day 2
X_train = pd.read_csv(os.path.join(DATA_DIR, 'X_train.csv'))
X_test  = pd.read_csv(os.path.join(DATA_DIR, 'X_test.csv'))
y_train = pd.read_csv(os.path.join(DATA_DIR, 'y_train.csv')).squeeze()
y_test  = pd.read_csv(os.path.join(DATA_DIR, 'y_test.csv')).squeeze()

print('X_train shape:', X_train.shape)
print('X_test shape :', X_test.shape)
print(f'Fraud in train: {y_train.mean()*100:.1f}%')
print(f'Fraud in test : {y_test.mean()*100:.1f}%')
```

---
## 2. Define the 3 Models
> We use `class_weight='balanced'` and `scale_pos_weight` to handle the 91.7% vs 8.3% imbalance.
> Without this, the model would just predict 'Not Fraud' for everything and look 91% accurate.

```python
# Calculate imbalance ratio for XGBoost
imbalance_ratio = (y_train == 0).sum() / (y_train == 1).sum()
print(f'Imbalance ratio (not fraud / fraud): {imbalance_ratio:.1f}')
print('This means for every 1 fraud case there are ~11 normal cases\n')

models = {
    'Logistic Regression': LogisticRegression(
        class_weight='balanced',  # automatically handles imbalance
        max_iter=2000,            # give it enough iterations to converge
        random_state=42
    ),
    'Random Forest': RandomForestClassifier(
        class_weight='balanced',  # handles imbalance
        n_estimators=100,         # 100 decision trees
        random_state=42
    ),
    'XGBoost': XGBClassifier(
        scale_pos_weight=imbalance_ratio,  # XGBoost's way of handling imbalance
        random_state=42,
        eval_metric='logloss'
    )
}

print('3 models ready to train:')
for name in models:
    print(f'  - {name}')
```

---
## 3. Train & Evaluate All 3 Models
> We train each model and measure 5 metrics:
> - **Accuracy**: overall correctness (least useful here due to imbalance)
> - **Precision**: of all predicted frauds, how many were actually fraud?
> - **Recall**: of all actual frauds, how many did we catch?
> - **F1-Score**: balance between precision and recall (most important metric here)
> - **ROC-AUC**: how well the model separates fraud from non-fraud overall

```python
results = []
trained_models = {}

for name, model in models.items():
    print(f'Training {name}...', end=' ')

    # Train
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]  # probability of fraud

    # Evaluate
    acc       = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall    = recall_score(y_test, y_pred)
    f1        = f1_score(y_test, y_pred)
    roc_auc   = roc_auc_score(y_test, y_prob)

    results.append({
        'Model': name,
        'Accuracy':  round(acc, 4),
        'Precision': round(precision, 4),
        'Recall':    round(recall, 4),
        'F1-Score':  round(f1, 4),
        'ROC-AUC':   round(roc_auc, 4)
    })

    trained_models[name] = (model, y_pred, y_prob)
    print('Done!')

print('\nAll models trained successfully!')
```

---
## 4. Comparison Table — All 3 Models


```python
results_df = pd.DataFrame(results).set_index('Model')

# Highlight the best score in each column
print('=== MODEL COMPARISON TABLE ===')
display(results_df.style
    .highlight_max(axis=0, color='lightgreen')
    .format('{:.4f}')
)

# Identify best model by F1-Score
best_model_name = results_df['F1-Score'].idxmax()
print(f'\n Best model by F1-Score: {best_model_name}')
print(f' F1-Score : {results_df.loc[best_model_name, "F1-Score"]}')
print(f' ROC-AUC  : {results_df.loc[best_model_name, "ROC-AUC"]}')
print('\n This model will be tuned tomorrow on Day 4.')
```

---
## 5. Plot — Comparison Bar Chart


```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
x = np.arange(len(metrics))
width = 0.25
colors = ['#4C72B0', '#DD8452', '#55A868']

fig, ax = plt.subplots(figsize=(12, 5))

for i, (model_name, color) in enumerate(zip(results_df.index, colors)):
    values = results_df.loc[model_name, metrics].values
    bars = ax.bar(x + i * width, values, width, label=model_name, color=color, alpha=0.85)
    for bar in bars:
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.003,
            f'{bar.get_height():.3f}',
            ha='center', va='bottom', fontsize=8
        )

ax.set_xticks(x + width)
ax.set_xticklabels(metrics)
ax.set_ylim(0, 1.12)
ax.set_ylabel('Score')
ax.set_title('Model Comparison — All Metrics', fontsize=13, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, 'plot6_model_comparison.png'), bbox_inches='tight')
plt.show()
```

---
## 6. Confusion Matrices — All 3 Models
> Shows exactly what each model gets right and wrong.
> **Top-left**: correctly predicted NOT fraud
> **Bottom-right**: correctly predicted FRAUD (what we care about)
> **Bottom-left**: missed frauds (False Negatives — dangerous!)
> **Top-right**: false alarms (False Positives)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, (name, (model, y_pred, y_prob)) in zip(axes, trained_models.items()):
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Fraud', 'Fraud'])
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    ax.set_title(name, fontsize=11, fontweight='bold')

plt.suptitle('Confusion Matrices — All 3 Models', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, 'plot7_confusion_matrices.png'), bbox_inches='tight')
plt.show()

print('Reading the matrix:')
print('  Top-left     = True Negatives  (correctly said NOT fraud)')
print('  Bottom-right = True Positives  (correctly caught fraud) <- want this HIGH')
print('  Bottom-left  = False Negatives (missed actual fraud)    <- want this LOW')
print('  Top-right    = False Positives (false alarms)')
```

---
## 7. ROC Curves — All 3 Models
> Shows how well each model separates fraud from non-fraud at different thresholds.
> The closer the curve is to the top-left corner, the better the model.

```python
fig, ax = plt.subplots(figsize=(8, 6))
colors = ['#4C72B0', '#DD8452', '#55A868']

for (name, (model, y_pred, y_prob)), color in zip(trained_models.items(), colors):
    auc = roc_auc_score(y_test, y_prob)
    RocCurveDisplay.from_predictions(
        y_test, y_prob,
        name=f'{name} (AUC={auc:.3f})',
        color=color, ax=ax
    )

ax.plot([0, 1], [0, 1], 'k--', label='Random (AUC=0.500)')
ax.set_title('ROC Curves — All 3 Models', fontsize=13, fontweight='bold')
ax.legend(loc='lower right')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, 'plot8_roc_curves.png'), bbox_inches='tight')
plt.show()
```

---
## 8. Detailed Report for Best Model

```python
best_model, best_y_pred, best_y_prob = trained_models[best_model_name]

print(f' DETAILED REPORT: {best_model_name} ')
print()
print(classification_report(y_test, best_y_pred, target_names=['Not Fraud', 'Fraud']))
print('Interpretation:')
print('  Precision (Fraud): when it predicts fraud, how often is it right?')
print('  Recall    (Fraud): of all real frauds, how many did it catch?')
print('  F1-Score  (Fraud): balance of precision and recall — your key metric')
```
